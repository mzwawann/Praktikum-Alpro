```python
gaji_per_jam = float(input("Masukkan gaji per jam: "))
jam_per_minggu = float(input("Masukkan jumlah jam kerja per minggu: "))
total_jam = jam_per_minggu * 5
pendapatan_kotor = gaji_per_jam * total_jam
pajak = 0.14 * pendapatan_kotor
pendapatan_bersih = pendapatan_kotor - pajak
pakaian = 0.10 * pendapatan_bersih
alat_tulis = 0.01 * pendapatan_bersih
sisa_uang = pendapatan_bersih - pakaian - alat_tulis
sedekah = 0.25 * sisa_uang
anak_yatim = 0.30 * sedekah
kaum_dhuafa = sedekah - anak_yatim

print("1. Pendapatan sebelum pajak       : Rp", int(pendapatan_kotor))
print("2. Pendapatan setelah pajak       : Rp", int(pendapatan_bersih))
print("3. Uang untuk pakaian & aksesoris : Rp", int(pakaian))
print("4. Uang untuk alat tulis          : Rp", int(alat_tulis))
print("5. Jumlah uang yang disedekahkan  : Rp", int(sedekah))
print("6. Untuk anak yatim               : Rp", int(anak_yatim))
print("7. Untuk kaum dhuafa              : Rp", int(kaum_dhuafa))
```
