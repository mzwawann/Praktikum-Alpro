```python
# Modal awal dan target
modal = 200_000_000
target_uang = 400_000_000

# Suku bunga per tahun
suku_bunga = 0.10

from math import log

# Menghitung waktu (dalam tahun)
waktu = log(target_uang / modal) / log(1 + suku_bunga)
print("Waktu yang dibutuhkan:", waktu, "tahun")

jumlah_tahun = int(waktu)
jumlah_bulan = (waktu - jumlah_tahun) * 12

jumlah_tahun, jumlah_bulan
```
